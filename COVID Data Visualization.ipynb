```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('./data/covid_19_data.csv')
df
```

```python
df_cases = df[['ObservationDate', 'Confirmed', 'Deaths', 'Recovered']]
df_grouped = df_cases.groupby('ObservationDate').sum()
df_grouped['Sick'] = df_grouped['Confirmed'] - df_grouped['Deaths'] - df_grouped['Recovered']
```

```python
df_grouped
```

```python
df_grouped.plot()
```
